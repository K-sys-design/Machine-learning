```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/IRIS.csv')
print(df.head())
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(6, 3))
    sns.scatterplot(x=df[feature], y=[0] * len(df), hue=df['species'], palette="deep")
    plt.title(f"Univariate Scatter Plot - {feature}")
    plt.xlabel(feature)
    plt.ylabel("Constant (0)")
    plt.legend(title="Species")
    plt.show()
```

```python

bivariate_pairs = [
    ('sepal_width', 'petal_width'),
    ('sepal_length', 'petal_length')
]

for x_feature, y_feature in bivariate_pairs:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df[x_feature], y=df[y_feature], hue=df['species'], palette="deep")
    plt.title(f"Bivariate Scatter Plot - {x_feature} vs {y_feature}")
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.legend(title="Species")
    plt.show()
```

```python

df_features = df.drop(columns=['Id'], errors='ignore')

sns.pairplot(df_features, hue="species", diag_kind="kde", palette="deep")
plt.suptitle("Multivariate Pairplot", y=1.02)
plt.show()
```
